```python
import sys
bold_start = "\033[1m"
bold_end = "\033[0;0m"

std_name=input("Enter Name or -1 to Exit:")
while std_name!="-1":
    std_id=input("Enter College ID:")
    
    num_grade=int(input("Enter number of Subjects or -1 to Exit:"))
    list_std_grades=[]
    X=0
    if num_grade==-1:
        print('K')
        #raise SystemExit
        sys.exit()
        
    while num_grade!=0:
        #list_std_grades.append(num_grade)
        std_grade=float(input("Enter marks for {0} Grade or -1 to Exit:".format(num_grade)))
        list_std_grades.append(std_grade)
        num_grade= num_grade - 1
    #list_std_grades.append(std_grade)
    
    grade_total=0
    while X!=len(list_std_grades):
        grade_total=grade_total+list_std_grades[X]
        X=X+1
    std_avg=grade_total/(X)
    print("Number of Grades : ",X)
    
    if std_avg>=70:
        std_status="Passed!"
    else:
        std_status="Failed"
    
    if std_avg>=90:
        std_grade="A"
    elif std_avg>=85:
        std_grade="B+"
    elif std_avg>=80:
        std_grade="B"
    elif std_avg>=75:
        std_grade="C+"
    elif std_avg>=70:
        std_grade="C"
    else:
        std_grade="F"
    print(" \n\n --====" + bold_start +"Student Report for {0}".format(std_name) + bold_end+"====--")
    print("\tStudent ID: ",std_id)
    #print("Student NAme:",std_name.title(),"\n")
    print("\tStudent Name:",std_name,"\n")
    
    x=0
    while x!=len(list_std_grades):
        print("\tGrade {0}: {1}".format(x+1,list_std_grades[x]))
        x=x+1
    print("\n\tCourse Status: ",std_status)
    print("\tCourse Average: ",round(std_avg,2))
    print("\tStudent Grade: ",std_grade)
    print("---====" + bold_start +"END OF STUDENT REPORT"+ bold_end+"====---")
    
    pgm_ctrl=input("\n\n Press Any Key to Exit....")
    std_name=input("Enter Name or -1 to Exit:")
```
